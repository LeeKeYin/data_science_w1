```python
import numpy as np
import random
```

```python
print("=== Part 1: NumPy Array Operations ===\n")
```

```python
#1.Create a numpy array of size 2 x 2
a=np.array([[1,2],[3,4]])
print(a)
```

```python
#2.Create a random numpy array of size 3 x 4
b=np.random.random((3,4))
print(b)
print(f"Shape: {b.shape}\n")
```

```python
#3.Create a numpy array of ones and a numpy array of zeroes
c=np.zeros((2,2))
d=np.ones((4,3))
print(c)
print(d)
```

```python
#4.Create a random numpy array of size 6x3 and transpose it
e=np.random.random((6,3))
print(e)
print(np.transpose(e))
```

```python
#5.Create two random numpy array of size n x p and p x m and then compute the dot product
n,p,m=2,3,4
f=np.random.random((n,p))
g=np.random.random((p,m))
print(f"f ({n}x{p}): ")
print(f)
print(f"\ng ({p}x{m}): ")
print(g)
result=np.dot(f,g)
print(f"\nDot Product Result ({n}x{m}): ")
print(result)
print(f"Dot product shape: {result.shape}")
```
